```python
image=cv2.imread('tigress.jpg')

cv2.imshow('tigressfi',image)
cv2.waitKey(0)         #is used to display an image in a window for a specified amount of time.
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```

```python
#image resize
image=cv2.imread('tigress.jpg')
img=cv2.resize(image,(500,750))
 
plt.imshow(img)
```

```python
image=cv2.imread('tigress.jpg')
img=cv2.resize(image,(500,300))

               
cv2.imshow('tigressf2',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#greyscale conversion

img=cv2.imread('tigress.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
```

```python
cv2.imshow('tigressf3',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#image save

img=cv2.imread('tigress.jpg',cv2.IMREAD_GRAYSCALE)
img=cv2.resize(image,(800,600))
cv2.imwrite("tigress1.jpg",img)
```
