```python
from tensorflow.keras.utils import img_to_array,array_to_img,load_img
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen=ImageDataGenerator(rotation_range=40,width_shift_range=0.2,height_shift_range=0.2,shear_range=0.2,
                          horizontal_flip=True,fill_mode='nearest')
```

* rotation_range=40:Randomly rotates the image within range 0-40
* width_shift_range=0.2:shifts the image horizontally
* height_shift_range=0.2:shifts the image vertically
* shear_range=0.2:shears the image in the counter clockwise direction
* horizontal_flip=True:Flips the image horizontally
* fill_mode='nearest':fills in newly created pixels after rotation or shifting with the nearest existing pixel's value
* image= loading and reshaping


```python
img=load_img('images.jpg')# this is a pill image
x=img_to_array(img)
x.shape
```

```python
x=x.reshape((1,)+x.shape)
```

```python
x.shape
```

```python
# the flow() command below generates batches of randomly transformed images
# and saves the results to the preview/directory
```

```python
i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir='image_dir',save_prefix='cl',save_format='jpeg'):
    i+=1
    if i>20:
        break    #otherwise the generator would loop indefinitely
```
